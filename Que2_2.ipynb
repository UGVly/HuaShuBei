```python
import pandas as pd

# Load the Excel file
file_path = r'C:\Users\21742\Desktop\华数杯\CODE\Q3YUCE2.xlsx'
data = pd.read_excel(file_path,index_col=0)

# Display the first few rows of the dataframe to understand its structure
data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams ['font.sans-serif']=['simHei']
# Checking for correlations
correlation_matrix = data.corr()

# Plotting the correlation matrix
plt.figure(figsize=(10,6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()

# Checking the distribution of the target variable (Electricity Production)
plt.figure(figsize=(10, 6))
sns.histplot(data['电力产量'], kde=True)
plt.title("Distribution of Electricity Production")
plt.xlabel("Electricity Production")
plt.ylabel("Frequency")
plt.show()
```
